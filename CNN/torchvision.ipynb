```python
## ref computervision cookbook
from torchvision import datasets
```

```python
path2data = "./data"
train_data = datasets.MNIST(path2data,train=True,download=True)
x_train,y_train = train_data.data,train_data.targets
print(x_train.shape)
print(y_train.shape)
```

```python
val_data = datasets.MNIST(path2data,train=False,download=True)
x_val,y_val = val_data.data,val_data.targets
print(x_val.shape)
print(y_val.shape)
```

```python
# add new dimention amd make it B*C*H*W
if len(x_train.shape) == 3:
    x_train = x_train.unsqueeze(1)
print(x_train.shape)

if len(x_val.shape) == 3:
    x_val = x_val.unsqueeze(1)
print(x_val.shape)
```

```python
from torchvision import utils
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def show(img):
    # convert tensor to numpy array
    npimg = img.numpy()
    # Convert to H*W*C shape
    npimg_tr=np.transpose(npimg, (1,2,0))
    plt.imshow(npimg_tr,interpolation='nearest')
```

```python
# make a grid of 40 images, 8 images per row
x_grid=utils.make_grid(x_train[:40], nrow=8, padding=2)
print(x_grid.shape)
# call helper function
show(x_grid)
```

```python
from torchvision import transforms
```

```python
train_data = datasets.MNIST(path2data,train=True,download=True)
```

```python
# transpose does cropping,resize etc.
data_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=1),
    transforms.RandomVerticalFlip(p =1),
    transforms.ToTensor(),
])
```

```python
img = train_data[0][0]
img_tr = data_transform(img)
img_tr_np = img_tr.numpy()
```

```python
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.title("original")
plt.subplot(1,2,2)
plt.imshow(img_tr_np[0],cmap="gray");
plt.title("transformed")
```
